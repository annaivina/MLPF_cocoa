```python
import json
import glob
import matplotlib.pyplot as plt
import re 
```

```python
def extract_number(filename):
    match = re.search(r'_(\d+)\.json', filename)
    if match:
        return int(match.group(1))
    else:
        return float('inf')  # Return a large number if no match is found

# Get a list of all JSON files
file_list = glob.glob('clusters/history_*.json')
```

```python
# Sort the file list based on the extracted number
sorted_files = sorted(file_list, key=extract_number)
```

```python
all_data = []
file_number = 1  # Initialize the counter

for file_path in sorted_files:
    with open(file_path, 'r') as file:
        # Load JSON data from the current file
        data = json.load(file)
        
        # Add the data to your list (or process it as needed)
        all_data.append(data)

```

```python
 for data in all_data:
     for key, value in data.items():     
         print(f"Key: {key}, Value: {value}")
```

```python
# Example keys
key1 = 'loss'
key2 = 'val_loss'

# Lists to hold the values
values_key1 = []
values_key2 = []

epochs = list(range(1, len(values_key1) + 1)) 
# Extract the values
for data in all_data:
    if key1 in data and key2 in data:
        values_key1.append(data[key1])
        values_key2.append(data[key2])
        
values_key1
# Now, plot the values
plt.figure(figsize=(10, 6))
plt.plot(values_key1, label=key1)
plt.plot(values_key2, label=key2)
plt.xticks(range(0, len(values_key1))) 
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()
```
